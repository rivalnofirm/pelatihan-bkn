```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://root:@127.0.0.1:3306/data_wilayah?charset=utf8mb4")
```

```python
query_wilayah ='''
        SELECT * FROM provinces
        ORDER BY provinsi_id'''

query_employee = '''
        SELECT * FROM employee'''
```

```python
data_wilayah = pd.read_sql(query_wilayah, engine)
data_employee = pd.read_sql(query_employee, engine)
```
