## Data Cleaning 

### Merubah nama column menggunakan fungsi .rename()

Fungsi .rename() beguna untuk merubah label pada suatu dataset, tujuan dari perubahan nama column ini berguna untuk mempermudah pada penamaan setiap column yang nantinya akan digunakan untuk tahapan explorasi data. Selain itu juga untuk menghilangkan beberapa karakter yang sekiranya tidak diperlukan pada nama column

```python
data_rename_column = data_selected.rename(columns={
    'Emp ID':'id_karyawan',
    'First Name':'nama_depan',
    'Last Name': 'nama_belakang',
    'Gender': 'jenis_kelamin',
    'E Mail': 'email',
    'Date of Birth':'tanggal_lahir',
    'Age in Yrs.': 'umur',
    'Date of Joining':'tanggal_gabung',
    'Year of Joining':'tahun_gabung',
    'Month Name of Joining':'bulan_gabung',
    'Day of Joining':'hari_gabung',
    'Salary':'gaji',
    'Last % Hike':'kenaikan_gaji',
    'Phone No. ':'nomer_telepon',
    'Province ID':'provinsi_id'
})
```

```python
data_rename_column.head()
```

```python
data_rename_column.info()
```

### Menggabungkan 2 column untuk menhasilkan column baru

Pada code dibawah ini, kita mencoba untuk mengabungkan 2 buah column untuk menghasilkan column baru. Disini akan dicontohkan untuk menggabungkan column nama_depan dan nama belakang untuk menhasilkan column nama

```python
data_rename_column['nama'] = data_rename_column[['nama_depan', 'nama_belakang']].apply(lambda x: ' '.join(x), axis=1)
```

```python
data_rename_column.head()
```

Setelah menggabungkan column nama_depan dan nama_belakang maka selanjutnya akan di lakukan lagi seleksi ulang pada data. Dimana column nama_depan dan nama_belakan tidak akan lg digunakan

```python
data_add_name = data_rename_column[[
    'id_karyawan',
    'nama',
    'jenis_kelamin',
    'email',
    'tanggal_lahir',
    'umur',
    'tanggal_gabung',
    'tahun_gabung',
    'bulan_gabung',
    'hari_gabung',
    'gaji',
    'kenaikan_gaji',
    'nomer_telepon',
    'provinsi_id'
]]
```

```python
data_add_name.info()
```

## Merubah tipe data pada column

merubah tipe data object menjadi datetime

```python
data_add_name["tanggal_lahir"] = pd.to_datetime(data_add_name["tanggal_lahir"])
data_add_name["tanggal_gabung"] = pd.to_datetime(data_add_name["tanggal_gabung"])
```

```python
data_add_name.info()
```

merubah tipe data object menjadi integer

```python
data_add_name['umur'] = data_add_name['umur'].astype(int)
```

```python
data_add_name.head()
```

```python
data_add_name.info()
```

menghilang kan karakter tertentu pada data, disini akan contohkan untuk menghilangka karakter % pada data dicolumn kenaikan gaji

```python
data_add_name["kenaikan_gaji"] = data_add_name["kenaikan_gaji"].str.replace("%", "")
```

```python
data_rename_col = data_add_name.rename(columns = {
    'kenaikan_gaji':'presentase_kenaikan_gaji'
})
```

```python
data_rename_col.head()
```

```python
data_rename_col['presentase_kenaikan_gaji'] = pd.to_numeric(data_rename_col['presentase_kenaikan_gaji'])
```

```python
data_rename_col.info()
```

## Data Aggeregation 

Pada code dibawah ini kita akan mencoba untuk mencari tahu berapa lama karyawan bekerja, kita akan membuat sebuah
column baru dengan nama lama _kerja. Dimana column lama bekerja ini adalah hasil pengurangan tahun saat ini di kurang
tahun awal karyawan tersebut bergabung.

```python
today = pd.to_datetime('today')
data_rename_col['lama_kerja'] = today.year - data_rename_col['tanggal_gabung'].dt.year
data_rename_col
```

Pada code ini kita juga akan coba berexplorasi dengan data yang ada. dimana kita akan membuat sebuah column untuk mendapatkan hasil
kenaikan_gaji_pertahun karyawan setiap tahunnya selama bekerja. rumus yang digunakan untuk mendapatkan hasil kenaikan gaji pertahun = gaji*presentase_kenaikan_gaji/100/lama_kerja

```python
data_rename_col['kenaikan_gaji_pertahun'] = data_rename_col['gaji']*data_rename_col['presentase_kenaikan_gaji']/100/data_rename_col['lama_kerja']
```

```python
data_rename_col.head()
```

```python
data_clean = data_rename_col[[
    'id_karyawan',
    'nama',
    'jenis_kelamin',
    'email',
    'tanggal_lahir',
    'umur',
    'tanggal_gabung',
    'tahun_gabung',
    'bulan_gabung',
    'hari_gabung',
    'lama_kerja',
    'gaji',
    'presentase_kenaikan_gaji',
    'kenaikan_gaji_pertahun',
    'nomer_telepon',
    'provinsi_id'
]]
```

```python
data_clean.head()
```

```python
data_clean['nama'] = data_clean.nama.str.lower()
```
