```python
data_selected = data[['Emp ID',
      'First Name',
      'Last Name',
      'Gender',
      'E Mail',
      'Date of Birth',
      'Age in Yrs.',
      'Date of Joining',
      'Year of Joining',
      'Month Name of Joining',
      'Day of Joining',
      'Salary',
      'Last % Hike',
      'Phone No. ',
      'Province ID']]
```

```python
data.head()
```

```python
data_selected.head()
```

```python
data_rename_column = data_selected.rename(columns={
    'Emp ID':'id_karyawan',
    'First Name':'nama_depan',
    'Last Name': 'nama_belakang',
    'Gender': 'jenis_kelamin',
    'E Mail': 'email',
    'Date of Birth':'tanggal_lahir',
    'Age in Yrs.': 'umur',
    'Date of Joining':'tanggal_gabung',
    'Year of Joining':'tahun_gabung',
    'Month Name of Joining':'bulan_gabung',
    'Day of Joining':'hari_gabung',
    'Salary':'gaji',
    'Last % Hike':'kenaikan_gaji',
    'Phone No. ':'nomer_telepon',
    'Province ID':'provinsi_id'
})
```

```python
data_rename_column.head()
```

```python
data_rename_column.info()
```

```python
data_rename_column['nama'] = data_rename_column[['nama_depan', 'nama_belakang']].apply(lambda x: ' '.join(x), axis=1)
```

```python
data_rename_column.head()
```

```python
data_add_name = data_rename_column[[
    'id_karyawan',
    'nama',
    'jenis_kelamin',
    'email',
    'tanggal_lahir',
    'umur',
    'tanggal_gabung',
    'tahun_gabung',
    'bulan_gabung',
    'hari_gabung',
    'gaji',
    'kenaikan_gaji',
    'nomer_telepon',
    'provinsi_id'
]]
```

```python
data_add_name.info()
```

```python
data_add_name["tanggal_lahir"] = pd.to_datetime(data_add_name["tanggal_lahir"])
data_add_name["tanggal_gabung"] = pd.to_datetime(data_add_name["tanggal_gabung"])
```

```python
data_add_name.info()
```

```python
data_add_name['umur'] = data_add_name['umur'].astype(int)
```

```python
data_add_name.head()
```

```python
data_add_name
```

```python
data_add_name.info()
```

```python
data_add_name["kenaikan_gaji"] = data_add_name["kenaikan_gaji"].str.replace("%", "")
```

```python
data_rename_col = data_add_name.rename(columns = {
    'kenaikan_gaji':'presentase_kenaikan_gaji'
})
```

```python
data_rename_col.head()
```

```python
data_rename_col['presentase_kenaikan_gaji'] = pd.to_numeric(data_rename_col['presentase_kenaikan_gaji'])
```

```python
data_rename_col.info()
```

```python
today = pd.to_datetime('today')
data_rename_col['lama_kerja'] = today.year - data_rename_col['tanggal_gabung'].dt.year
data_rename_col
```

```python
data_rename_col['kenaikan_gaji_pertahun'] = data_rename_col['gaji']*data_rename_col['presentase_kenaikan_gaji']/100/data_rename_col['hari_gabung']
```

```python
data_rename_col.head()
```

```python
data_clean = data_rename_col[[
    'id_karyawan',
    'nama',
    'jenis_kelamin',
    'email',
    'tanggal_lahir',
    'umur',
    'tanggal_gabung',
    'tahun_gabung',
    'bulan_gabung',
    'hari_gabung',
    'lama_kerja',
    'gaji',
    'presentase_kenaikan_gaji',
    'kenaikan_gaji_pertahun',
    'nomer_telepon',
    'provinsi_id'
]]
```

```python
data_clean.head()
```

```python
data_clean['nama'] = data_clean.nama.str.lower()
```
